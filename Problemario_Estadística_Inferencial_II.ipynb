# Unidad 1. Regresión lineal simple y correlación

## Medidas de dispersión
Suma de los cuadrados de las desviaciones de los valores de $X$ con respecto a su media:
$$ S_{xx} = \sum_{i=1}^{n} (X_i - \bar{X})^2 $$
Suma de los productos de las desviaciones de los valores de $X$ y $Y$ con respecto a sus medias:
$$S_{xy} = \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})$$
Suma de los cuadrados de las desviaciones de los valores de $Y$ con respecto a su media:
$$ S_{yy} = \sum_{i=1}^{n} (Y_i - \bar{Y})^2 $$


## Coeficiente de correlación y de determinación
Coeficiente de correlación de Pearson:
$$ r = \frac{S_{xy}}{\sqrt{S_{xx} \cdot S_{yy}}} $$
Coeficiente de determinación:
$$ r^2 $$

## Recta de regresión ajustada
La regresión lineal ajustada se representa mediante estadísticos:
$$ \hat{Y} = b_0 + b_1 X $$
donde $\hat{Y}$ representa el valor de $Y$ obtenido mediante la recta de regresión ajustada (no la verdadera $Y$). Los estadísticos $b_0$ y $b_1$ se calculan de la siguiente manera:

$$b_1 = \frac{S_{xy}}{S_{xx}}$$

$$ b_0 = \bar{Y} - b_1 \bar{X} $$

## Cálculo de residuales
Residuales:
$$e_i = Y_i - \hat{Y}_i$$

## Sumas de cuadrados SS (Sum of Squares)
Suma de los Cuadrados de los Errores:
$$ SSE = \sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2 = \sum_{i=1}^n e_i^2$$
Suma total de cuadrados:
$$SST = \sum_{i=1}^{n} (Y_i - \bar{Y})^2$$
Suma de cuadrados de regresión:
$$SSR = \sum_{i=1}^{n} (\hat{Y_i} - \bar{Y})^2$$
- SST: Mide la variabilidad total de los datos observados.
- SSR: Mide la variabilidad de los datos que el modelo de regresión explica.
- SSE: Mide la variabilidad no explicada por el modelo (es decir, los residuos).


## Intervalo de confianza
Estadístico de prueba _t_:
$$t = \frac{b_1}{SE(b_1)}$$
Error estándar de $b_1$:
$$SE(b_1) = \frac{\sqrt{SSE / (n-2)}}   {\sqrt{S_{xx}} }$$
o
$$SE(b_1) = \frac{\sqrt{\sum_{i=1}^{n} (Y_i - \hat{Y}_i)^2  / (n-2)}}   {\sqrt{S_{xx}} }$$

Intervalo de confianza para $b_1$:
$$b_1 - t_{\alpha/2} \cdot SE(b_1) \lt \beta_1 \lt b_1 + t_{\alpha/2} \cdot SE(b_1)$$
donde $n$ representa la cantidad de pares de datos.

## Comprobación de supuestos
Comprobar suposiciones:
- Test de shapiro a los residuales $e_i$: Para comprobar si la distribución es normal sobre la recta.
- Grafico $X$ vs $Y$: Para observar si los datos soportan la suposición de linealidad.
- Gráfico de residuales: Para observar si los datos soportan la suposición de linealidad, complementario al coeficiente de correlación
- Test de Breusch-Pagan: Para detectar heteroscedasticidad en regresion lineal

Test de Shapiro:<br>
`from scipy.stats import shapiro`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_sh = shapiro(data)`<br>
- $H_0$: Los datos siguen una distribución normal
- $H_1$: Los datos no siguen una distribución normal

Test de Breusch-Pagan:<br>
`from statsmodels.stats.api import het_breuschpagan`<br>
Después, se obtiene el valor-p:<br>
`_, valor_p_bp, _, _ = het_breuschpagan(residuales, X)`
- $H_0$: Hay homoscedasticidad
- $H_1$: Hay heteroscedasticidad

## ANOVA en regresión lineal

| Fuente de variación | Suma de cuadrados (SS) | Grados de libertad (df) | Promedio de los cuadrados (MS) | Estadístico F |
|---------------------|------------------------|--------------------------|--------------------------------|---------------|
| Regresión  | $ SSR$ | $p$ | $MSR = \frac{SSR}{p}$ | $F = \frac{MSR}{MSE}$|
| Error | $SSE$| $n - p - 1 $| $MSE = \frac{SSE}{n - p - 1} $|     |
| Total | $SST $| $n - 1$|             |        |

donde $p$ es el número de parámetros para la recta de regresión ajustada (en la regresión simple $p=1$). Las hipótesis son:
$$H_0: \beta_1 = 0$$
$$H_1: \beta_1 \ne 0 $$

### Problema 1
Un profesor intenta mostrar a sus estudiantes la importancia de los exámenes cortos, aun cuando el 90% de la calificación final esté determinada por los exámenes parciales. Él cree que cuanto más altas sean las calificaciones de los exámenes cortos, más alta será la calificación final. Seleccionó una muestra aleatoria de 15 estudiantes de su clase con los siguientes datos:

| Promedio de exámenes cortos | Promedio final |
|--------------|--------------|
| 59 | 64 |
| 92 | 84 |
| 72 | 77 |
| 90 | 80 |
| 95 | 77 |
| 87 | 81 |
| 89 | 80 |
| 77 | 84 |
| 76 | 80 |
| 65 | 69 |
| 97 | 83 |
| 42 | 40 |
| 94 | 78 |
| 62 | 65 |
| 91 | 90 |



1.   Establesca una variable dependiente ($Y$) y una variable independiente ($X$).

Variable dependiente: Promedio Final.
Variable independiente: Examenes cortos.

2.   Realice un diagrama de dispersión para estos datos.
3. ¿Los datos soportan la suposición de linealidad?
Si
4. Calcule el coeficiente de correlación e interprete el resultado:0.8646
5. Calcule el coeficiente de determinación e interprete el resultado:
 74.75%
6. Obtenga la recta de regresión ajustada y grafíquelo sobre el gráfico de dispersión.
7. Obtenga un intervalo de confianza del 95% para la pendiente de la recta de regresión ajustada ($b_1$)
8. Calcule los residuales y trace un nuevo gráfico de dispersión. Comente, ¿Parece que se verifican los supuestos?: No
9. Realice la prueba de Shapiro para los residuales y comente el resultado: 0.901827735700704
10. Realice la prueba de Brausch-Pagan para los residuales y comente el resultado: 0.2289
11. Tres estudiantes sacaron 70, 75 y 84 de calificación. Según la recta de regresión ajustada, ¿cuáles son los resultados esperados para estos tres alumnos?: 78-80 de calificación
12. Realice una tabla ANOVA e interprete el resultado.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos del Problema 1
data = {
    'Promedio de exámenes cortos': [59, 92, 72, 90, 95, 87, 89, 77, 76, 65, 97, 42, 94, 62, 91],
    'Promedio final': [64, 84, 77, 80, 77, 81, 80, 84, 80, 69, 83, 40, 78, 65, 90]
}
df_problema1 = pd.DataFrame(data)

# Establecer variables dependiente (Y) e independiente (X)
X = df_problema1['Promedio de exámenes cortos']
Y = df_problema1['Promedio final']

# 2. Realizar un diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X, y=Y, s=100, alpha=0.8)
plt.title('Diagrama de Dispersión: Promedio de Exámenes Cortos vs. Promedio Final')
plt.xlabel('Promedio de Exámenes Cortos (X)')
plt.ylabel('Promedio Final (Y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 3. ¿Los datos soportan la suposición de linealidad?
# Se puede comentar visualmente sobre la linealidad a partir del gráfico.
# En este caso, los puntos parecen seguir una tendencia lineal ascendente, lo que sugiere una relación lineal.
print("Visualmente, los datos parecen soportar la suposición de linealidad, mostrando una tendencia ascendente.")
```

```python
import numpy as np

# Cálculo de las medidas de dispersión
mean_X = np.mean(X)
mean_Y = np.mean(Y)

S_xx = np.sum((X - mean_X)**2)
S_yy = np.sum((Y - mean_Y)**2)
S_xy = np.sum((X - mean_X) * (Y - mean_Y))

# 4. Coeficiente de correlación de Pearson (r)
r = S_xy / np.sqrt(S_xx * S_yy)
print(f"Coeficiente de correlación (r): {r:.4f}")

# Interpretación del coeficiente de correlación
if r > 0.7:
    corr_interpretation = "una fuerte correlación lineal positiva"
elif r > 0.3:
    corr_interpretation = "una correlación lineal positiva moderada"
elif r > 0:
    corr_interpretation = "una correlación lineal positiva débil"
elif r < -0.7:
    corr_interpretation = "una fuerte correlación lineal negativa"
elif r < -0.3:
    corr_interpretation = "una correlación lineal negativa moderada"
elif r < 0:
    corr_interpretation = "una correlación lineal negativa débil"
else:
    corr_interpretation = "muy poca o ninguna correlación lineal"

print(f"Interpretación: Existe {corr_interpretation} entre el promedio de exámenes cortos y el promedio final.")

# 5. Coeficiente de determinación (r^2)
r_squared = r**2
print(f"Coeficiente de determinación (r^2): {r_squared:.4f}")

# Interpretación del coeficiente de determinación
print(f"Interpretación: El {r_squared*100:.2f}% de la variabilidad del promedio final (Y) puede explicarse por la variabilidad del promedio de exámenes cortos (X).")
```

```python
# 6. Obtener la recta de regresión ajustada
b1 = S_xy / S_xx
b0 = mean_Y - b1 * mean_X

print(f"Estadístico b1 (pendiente): {b1:.4f}")
print(f"Estadístico b0 (intersección): {b0:.4f}")
print(f"Ecuación de la recta de regresión ajustada: Y_hat = {b0:.4f} + {b1:.4f}X")

# Graficar la recta de regresión sobre el diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X, y=Y, s=100, alpha=0.8)
plt.plot(X, b0 + b1 * X, color='green', linewidth=2, label=f'Y_hat = {b0:.2f} + {b1:.2f}X')
plt.title('Diagrama de Dispersión con Recta de Regresión Ajustada')
plt.xlabel('Promedio de Exámenes Cortos (X)')
plt.ylabel('Promedio Final (Y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

### Cálculo y Gráfico de Residuales
Los **residuales** ($e_i$) son la diferencia entre el valor observado de la variable dependiente ($Y_i$) y el valor predicho por el modelo de regresión ($\hat{Y}_i$). Es decir:

$$e_i = Y_i - \hat{Y}_i$$

Un gráfico de dispersión de los residuales contra la variable independiente ($X$) o los valores predichos ($\hat{Y}_i$) es una herramienta útil para verificar los supuestos de la regresión lineal, como la linealidad, la homocedasticidad y la normalidad de los errores.

```python
# Calcular los valores predichos (Y_hat)
Y_hat = b0 + b1 * X

# Calcular los residuales
residuales = Y - Y_hat

# 8. Calcular los residuales y trazar un nuevo gráfico de dispersión.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X, y=residuales, s=100, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--') # Añadir una línea en y=0 para referencia
plt.title('Gráfico de Dispersión de Residuales vs. Promedio de Exámenes Cortos')
plt.xlabel('Promedio de Exámenes Cortos (X)')
plt.ylabel('Residuales ($e_i$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Comentario sobre el gráfico de residuales
print("Comentario: En este gráfico de residuales, los puntos parecen dispersarse aleatoriamente alrededor de la línea cero, sin mostrar un patrón discernible. Esto sugiere que los supuestos de linealidad y homocedasticidad podrían ser válidos. No se observan patrones en forma de cono (heterocedasticidad) ni curvas (no linealidad).")
```

```python
from scipy.stats import shapiro

# 9. Realice la prueba de Shapiro para los residuales
statistic, p_value_shapiro = shapiro(residuales)

print(f"Estadístico de Shapiro-Wilk: {statistic:.4f}")
print(f"Valor p de Shapiro-Wilk: {p_value_shapiro:.4f}")

# Interpretación del resultado de la prueba de Shapiro-Wilk
alpha = 0.05
if p_value_shapiro > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Los residuales parecen seguir una distribución normal.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Los residuales no parecen seguir una distribución normal.")
```

```python
from statsmodels.stats.api import het_breuschpagan
import statsmodels.api as sm

# 10. Realice la prueba de Breusch-Pagan para los residuales
# El test de Breusch-Pagan requiere la variable independiente (X) en formato de array 2D y un término constante.
X_bp_no_constant = X.values.reshape(-1, 1)
X_bp = sm.add_constant(X_bp_no_constant) # Añadir una constante

# Realizar la prueba de Breusch-Pagan
# Retorna: lm_test, f_test, bp_test, jb_test
# Solo nos interesa el p-value del test de Breusch-Pagan
lm_test, lm_p_value, f_test, f_p_value = het_breuschpagan(residuales, X_bp)

print(f"Estadístico de Lagrange Multiplier (LM): {lm_test:.4f}")
print(f"Valor p de la prueba LM: {lm_p_value:.4f}")

# Interpretación del resultado de la prueba de Breusch-Pagan
alpha = 0.05
if lm_p_value > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Hay evidencia de homocedasticidad en los residuales.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Hay evidencia de heterocedasticidad en los residuales.")
```

```python
# 11. Predicciones para nuevos valores de X
nuevos_promedios_cortos = pd.Series([70, 75, 84])
predicciones_Y = b0 + b1 * nuevos_promedios_cortos

print("Predicciones de promedio final para los nuevos alumnos:")
for i, promedio_corto in enumerate(nuevos_promedios_cortos):
    print(f"  Promedio de exámenes cortos: {promedio_corto} -> Promedio final esperado: {predicciones_Y.iloc[i]:.2f}")

print("\nComentario: Estos resultados representan la calificación final esperada según nuestro modelo de regresión lineal. La interpolación (para 70 y 75) se considera más fiable ya que está dentro del rango de los datos observados, mientras que la extrapolación (para 84, si 84 está fuera del rango máximo de X original, aunque en este caso 84 está dentro) podría tener menos precisión.")
```

```python
import numpy as np
from scipy.stats import f

# 12. Realice una tabla ANOVA e interprete el resultado.

# Número de observaciones
n = len(Y)

# Sumas de cuadrados (ya calculadas en parte o se pueden recalcular)
# SST = S_yy (ya calculado)
SST = S_yy

# SSR = Suma de cuadrados de regresión
SSR = np.sum((Y_hat - mean_Y)**2)

# SSE = Suma de cuadrados de los errores
SSE = np.sum(residuales**2)

# Grados de libertad
df_regression = 1 # Para regresión lineal simple (p=1)
df_error = n - 2
df_total = n - 1

# Promedios de los cuadrados
MSR = SSR / df_regression
MSE = SSE / df_error

# Estadístico F
F_statistic = MSR / MSE

# Valor p del estadístico F
p_value_anova = 1 - f.cdf(F_statistic, df_regression, df_error)

print("### Tabla ANOVA para el Modelo de Regresión ###")
print("------------------------------------------------------------------------------------------------")
print(f"| {'Fuente de variación':<20} | {'Suma de cuadrados (SS)':<25} | {'Grados de libertad (df)':<25} | {'Promedio de los cuadrados (MS)':<28} | {'Estadístico F':<15} |")
print("------------------------------------------------------------------------------------------------")
print(f"| {'Regresión':<20} | {SSR:<25.4f} | {df_regression:<25} | {MSR:<28.4f} | {F_statistic:<15.4f} |")
print(f"| {'Error':<20} | {SSE:<25.4f} | {df_error:<25} | {MSE:<28.4f} | {'':<15} |")
print(f"| {'Total':<20} | {SST:<25.4f} | {df_total:<25} | {'':<28} | {'':<15} |")
print("------------------------------------------------------------------------------------------------")
print(f"Valor p del Estadístico F: {p_value_anova:.4f}")

# Interpretación de la tabla ANOVA
alpha = 0.05
print("\nInterpretación de la tabla ANOVA:")
if p_value_anova < alpha:
    print(f"Dado que el valor p ({p_value_anova:.4f}) es menor que el nivel de significancia (alpha={alpha}), rechazamos la hipótesis nula (H0: β1 = 0).")
    print("Esto significa que existe una relación lineal estadísticamente significativa entre el promedio de exámenes cortos y el promedio final.")
    print("El modelo de regresión es significativo para predecir el promedio final.")
else:
    print(f"Dado que el valor p ({p_value_anova:.4f}) es mayor que el nivel de significancia (alpha={alpha}), no rechazamos la hipótesis nula (H0: β1 = 0).")
    print("Esto significa que no hay suficiente evidencia para afirmar que existe una relación lineal estadísticamente significativa entre el promedio de exámenes cortos y el promedio final.")
    print("El modelo de regresión no es significativo para predecir el promedio final.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos del Problema 2
data_problema2 = {
    'Categoría de salario': [11, 10, 8, 5, 9, 7, 3, 11, 8, 7, 2, 9, 8, 3],
    'Ausencias': [18, 17, 29, 36, 11, 28, 35, 14, 20, 32, 39, 16, 31, 40]
}
df_problema2 = pd.DataFrame(data_problema2)

# 1. Establecer variables dependiente (Y) e independiente (X)
# Según el problema: Salario (Y) es dependiente, Ausencias (X) es independiente.
X_p2 = df_problema2['Ausencias']
Y_p2 = df_problema2['Categoría de salario']

# 2. Realizar un diagrama de dispersión
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_p2, y=Y_p2, s=100, alpha=0.8)
plt.title('Diagrama de Dispersión: Ausencias vs. Categoría de Salario')
plt.xlabel('Ausencias (X)')
plt.ylabel('Categoría de Salario (Y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 3. ¿Los datos soportan la suposición de linealidad?
print("Visualmente, los datos parecen mostrar una tendencia negativa, lo que podría indicar una relación lineal inversa. Sin embargo, hay cierta dispersión, por lo que es importante cuantificarla.")
```

```python
import numpy as np

# Cálculo de las medidas de dispersión para el Problema 2
mean_X_p2 = np.mean(X_p2)
mean_Y_p2 = np.mean(Y_p2)

S_xx_p2 = np.sum((X_p2 - mean_X_p2)**2)
S_yy_p2 = np.sum((Y_p2 - mean_Y_p2)**2)
S_xy_p2 = np.sum((X_p2 - mean_X_p2) * (Y_p2 - mean_Y_p2))

# 4. Coeficiente de correlación de Pearson (r) para el Problema 2
r_p2 = S_xy_p2 / np.sqrt(S_xx_p2 * S_yy_p2)
print(f"Coeficiente de correlación (r) para el Problema 2: {r_p2:.4f}")

# Interpretación del coeficiente de correlación para el Problema 2
if r_p2 > 0.7:
    corr_interpretation_p2 = "una fuerte correlación lineal positiva"
elif r_p2 > 0.3:
    corr_interpretation_p2 = "una correlación lineal positiva moderada"
elif r_p2 > 0:
    corr_interpretation_p2 = "una correlación lineal positiva débil"
elif r_p2 < -0.7:
    corr_interpretation_p2 = "una fuerte correlación lineal negativa"
elif r_p2 < -0.3:
    corr_interpretation_p2 = "una correlación lineal negativa moderada"
elif r_p2 < 0:
    corr_interpretation_p2 = "una correlación lineal negativa débil"
else:
    corr_interpretation_p2 = "muy poca o ninguna correlación lineal"

print(f"Interpretación: Existe {corr_interpretation_p2} entre Ausencias y Categoría de Salario.")
```

```python
# 5. Coeficiente de determinación (r^2) para el Problema 2
r_squared_p2 = r_p2**2
print(f"Coeficiente de determinación (r^2) para el Problema 2: {r_squared_p2:.4f}")

# Interpretación del coeficiente de determinación para el Problema 2
print(f"Interpretación: El {r_squared_p2*100:.2f}% de la variabilidad de la Categoría de Salario (Y) puede explicarse por la variabilidad de las Ausencias (X) en el Problema 2.")
```

```python
# 6. Obtener la recta de regresión ajustada para el Problema 2
b1_p2 = S_xy_p2 / S_xx_p2
b0_p2 = mean_Y_p2 - b1_p2 * mean_X_p2

print(f"Estadístico b1 (pendiente) para el Problema 2: {b1_p2:.4f}")
print(f"Estadístico b0 (intersección) para el Problema 2: {b0_p2:.4f}")
print(f"Ecuación de la recta de regresión ajustada para el Problema 2: Y_hat_p2 = {b0_p2:.4f} + {b1_p2:.4f}X")

# Graficar la recta de regresión sobre el diagrama de dispersión para el Problema 2
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_p2, y=Y_p2, s=100, alpha=0.8)
plt.plot(X_p2, b0_p2 + b1_p2 * X_p2, color='red', linewidth=2, label=f'Y_hat_p2 = {b0_p2:.2f} + {b1_p2:.2f}X')
plt.title('Diagrama de Dispersión con Recta de Regresión Ajustada (Problema 2)')
plt.xlabel('Ausencias (X)')
plt.ylabel('Categoría de Salario (Y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
# Calcular los valores predichos (Y_hat) para el Problema 2
Y_hat_p2 = b0_p2 + b1_p2 * X_p2

# Calcular los residuales para el Problema 2
residuales_p2 = Y_p2 - Y_hat_p2

# 8. Calcular los residuales y trazar un nuevo gráfico de dispersión.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_p2, y=residuales_p2, s=100, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--') # Añadir una línea en y=0 para referencia
plt.title('Gráfico de Dispersión de Residuales vs. Ausencias (Problema 2)')
plt.xlabel('Ausencias (X)')
plt.ylabel('Residuales ($e_i$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Comentario sobre el gráfico de residuales para el Problema 2
print("Comentario: Observando el gráfico de residuales, buscaremos patrones que puedan indicar violaciones de los supuestos. Una dispersión aleatoria alrededor de la línea cero sugiere linealidad y homocedasticidad. Patrones en forma de cono indicarían heterocedasticidad, y patrones curvos, no linealidad.")
```

```python
from scipy.stats import shapiro

# 9. Realice la prueba de Shapiro para los residuales del Problema 2
statistic_p2, p_value_shapiro_p2 = shapiro(residuales_p2)

print(f"Estadístico de Shapiro-Wilk para el Problema 2: {statistic_p2:.4f}")
print(f"Valor p de Shapiro-Wilk para el Problema 2: {p_value_shapiro_p2:.4f}")

# Interpretación del resultado de la prueba de Shapiro-Wilk para el Problema 2
alpha = 0.05
if p_value_shapiro_p2 > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Los residuales del Problema 2 parecen seguir una distribución normal.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Los residuales del Problema 2 no parecen seguir una distribución normal.")
```

```python
from statsmodels.stats.api import het_breuschpagan
import statsmodels.api as sm

# 10. Realice la prueba de Breusch-Pagan para los residuales del Problema 2
# El test de Breusch-Pagan requiere la variable independiente (X) en formato de array 2D y un término constante.
X_bp_p2_no_constant = X_p2.values.reshape(-1, 1)
X_bp_p2 = sm.add_constant(X_bp_p2_no_constant) # Añadir una constante

# Realizar la prueba de Breusch-Pagan
lm_test_p2, lm_p_value_p2, f_test_p2, f_p_value_p2 = het_breuschpagan(residuales_p2, X_bp_p2)

print(f"Estadístico de Lagrange Multiplier (LM) para el Problema 2: {lm_test_p2:.4f}")
print(f"Valor p de la prueba LM para el Problema 2: {lm_p_value_p2:.4f}")

# Interpretación del resultado de la prueba de Breusch-Pagan para el Problema 2
alpha = 0.05
if lm_p_value_p2 > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Hay evidencia de homocedasticidad en los residuales del Problema 2.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Hay evidencia de heterocedasticidad en los residuales del Problema 2.")
```

```python
import numpy as np

# Cálculo de las medidas de dispersión para el Problema 4
mean_X_p4 = np.mean(X_p4)
mean_Y_p4 = np.mean(Y_p4)

S_xx_p4 = np.sum((X_p4 - mean_X_p4)**2)
S_yy_p4 = np.sum((Y_p4 - mean_Y_p4)**2)
S_xy_p4 = np.sum((X_p4 - mean_X_p4) * (Y_p4 - mean_Y_p4))

# 4. Coeficiente de correlación de Pearson (r) para el Problema 4
r_p4 = S_xy_p4 / np.sqrt(S_xx_p4 * S_yy_p4)
print(f"Coeficiente de correlación (r) para el Problema 4: {r_p4:.4f}")

# Interpretación del coeficiente de correlación para el Problema 4
if r_p4 > 0.7:
    corr_interpretation_p4 = "una fuerte correlación lineal positiva"
elif r_p4 > 0.3:
    corr_interpretation_p4 = "una correlación lineal positiva moderada"
elif r_p4 > 0:
    corr_interpretation_p4 = "una correlación lineal positiva débil"
elif r_p4 < -0.7:
    corr_interpretation_p4 = "una fuerte correlación lineal negativa"
elif r_p4 < -0.3:
    corr_interpretation_p4 = "una correlación lineal negativa moderada"
elif r_p4 < 0:
    corr_interpretation_p4 = "una correlación lineal negativa débil"
else:
    corr_interpretation_p4 = "muy poca o ninguna correlación lineal"

print(f"Interpretación: Existe {corr_interpretation_p4} entre la construcción de casas y las ventas de línea blanca.")
```

```python
# 5. Coeficiente de determinación (r^2) para el Problema 4
r_squared_p4 = r_p4**2
print(f"Coeficiente de determinación (r^2) para el Problema 4: {r_squared_p4:.4f}")

# Interpretación del coeficiente de determinación para el Problema 4
print(f"Interpretación: El {r_squared_p4*100:.2f}% de la variabilidad de las Ventas de línea blanca (Y) puede explicarse por la variabilidad de la Construcción de casas (X) en el Problema 4.")
```

```python
# 6. Obtener la recta de regresión ajustada para el Problema 4
b1_p4 = S_xy_p4 / S_xx_p4
b0_p4 = mean_Y_p4 - b1_p4 * mean_X_p4

print(f"Estadístico b1 (pendiente) para el Problema 4: {b1_p4:.4f}")
print(f"Estadístico b0 (intersección) para el Problema 4: {b0_p4:.4f}")
print(f"Ecuación de la recta de regresión ajustada para el Problema 4: Y_hat_p4 = {b0_p4:.4f} + {b1_p4:.4f}X")

# Graficar la recta de regresión sobre el diagrama de dispersión para el Problema 4
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_p4, y=Y_p4, s=100, alpha=0.8)
plt.plot(X_p4, b0_p4 + b1_p4 * X_p4, color='purple', linewidth=2, label=f'Y_hat_p4 = {b0_p4:.2f} + {b1_p4:.2f}X')
plt.title('Diagrama de Dispersión con Recta de Regresión Ajustada (Problema 4)')
plt.xlabel('Construcción de Casas (miles) (X)')
plt.ylabel('Ventas de Línea Blanca (miles) (Y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular los valores predichos (Y_hat) para el Problema 4
Y_hat_p4 = b0_p4 + b1_p4 * X_p4

# Calcular los residuales para el Problema 4
residuales_p4 = Y_p4 - Y_hat_p4

# 8. Calcular los residuales y trazar un nuevo gráfico de dispersión.
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_p4, y=residuales_p4, s=100, alpha=0.8)
plt.axhline(y=0, color='r', linestyle='--') # Añadir una línea en y=0 para referencia
plt.title('Gráfico de Dispersión de Residuales vs. Construcción de Casas (Problema 4)')
plt.xlabel('Construcción de Casas (miles) (X)')
plt.ylabel('Residuales ($e_i$)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# Comentario sobre el gráfico de residuales para el Problema 4
print("Comentario: Observando el gráfico de residuales, buscaremos patrones que puedan indicar violaciones de los supuestos. Una dispersión aleatoria alrededor de la línea cero sugiere linealidad y homocedasticidad. Patrones en forma de cono indicarían heterocedasticidad, y patrones curvos, no linealidad.")
```

```python
from scipy.stats import shapiro

# 9. Realice la prueba de Shapiro para los residuales del Problema 4
statistic_p4, p_value_shapiro_p4 = shapiro(residuales_p4)

print(f"Estadístico de Shapiro-Wilk para el Problema 4: {statistic_p4:.4f}")
print(f"Valor p de Shapiro-Wilk para el Problema 4: {p_value_shapiro_p4:.4f}")

# Interpretación del resultado de la prueba de Shapiro-Wilk para el Problema 4
alpha = 0.05
if p_value_shapiro_p4 > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Los residuales del Problema 4 parecen seguir una distribución normal.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Los residuales del Problema 4 no parecen seguir una distribución normal.")
```

```python
from statsmodels.stats.api import het_breuschpagan
import statsmodels.api as sm

# 10. Realice la prueba de Breusch-Pagan para los residuales del Problema 4
# El test de Breusch-Pagan requiere la variable independiente (X) en formato de array 2D y un término constante.
X_bp_p4_no_constant = X_p4.values.reshape(-1, 1)
X_bp_p4 = sm.add_constant(X_bp_p4_no_constant) # Añadir una constante

# Realizar la prueba de Breusch-Pagan
lm_test_p4, lm_p_value_p4, f_test_p4, f_p_value_p4 = het_breuschpagan(residuales_p4, X_bp_p4)

print(f"Estadístico de Lagrange Multiplier (LM) para el Problema 4: {lm_test_p4:.4f}")
print(f"Valor p de la prueba LM para el Problema 4: {lm_p_value_p4:.4f}")

# Interpretación del resultado de la prueba de Breusch-Pagan para el Problema 4
alpha = 0.05
if lm_p_value_p4 > alpha:
    print("Comentario: El valor p es mayor que el nivel de significancia (alpha=0.05). No rechazamos la hipótesis nula. Hay evidencia de homocedasticidad en los residuales del Problema 4.")
else:
    print("Comentario: El valor p es menor o igual que el nivel de significancia (alpha=0.05). Rechazamos la hipótesis nula. Hay evidencia de heterocedasticidad en los residuales del Problema 4.")
```
